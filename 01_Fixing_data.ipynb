```python
# Describing data. There are outliers. It is necesary to delete them.

final_dataset_14.columns
```

```python
# Procedure to fill gaps using mean between previus and next data.

def fill_gap (dataset, columns_to_fill):
    
    # Fill NaN interpolating between before and after values
    
    try:
        dataset = dataset.interpolate(method='linear')
    except:
        dataset = dataset.fillna(method='bfill', limit=1)
    
    for nam_col in list(columns_to_fill):
        
        # Reference to considere outliers.
        
        ref_low = dataset[nam_col].mean() - dataset[nam_col].std() * 3
        ref_high = dataset[nam_col].mean() + dataset[nam_col].std() * 3
        
        # Saving index of outliers.
        
        indexes = dataset.index[(dataset[nam_col]<ref_low) & (dataset[nam_col]<ref_high)]
        
        # Changing outliers for a regular data. Taking care with extreme indexes.
        for index in indexes:
                
            if 0 < index < len(dataset):
                dataset.loc[index, nam_col] = (np.mean([dataset.loc[index-1, nam_col], 
                                                        dataset.loc[index+1, nam_col]]))   
            elif index == 0:
                dataset.loc[index, nam_col] = dataset.loc[(index+1), nam_col]
                
            elif index == len(dataset):
                dataset.loc[index, nam_col] = dataset.loc[index-1, nam_col]

    # If there are nay NaN, drop row
    
    dataset = dataset.dropna(how='any')
    
    return dataset
```

```python
# Applying procedure

final_dataset_14 = fill_gap (final_dataset_14, ['Ambient_Temperature', 'Global_Radiation',
       'Diffuse_Radiation', 'Ultraviolet', 'Wind_Velocity',
       'A_Vertical_VDC', 'A_Vertical_CDC', 'A_Vertical_PDC', 'A_Optimal_VDC',
       'A_Optimal_CDC', 'A_Optimal_PDC', 'A_Horizontal_VDC',
       'A_Horizontal_CDC', 'A_Horizontal_PDC', 'A_Vertical_Temperature',
       'A_Optimal_Temperature', 'A_Horizontal_Temperature', 'B_Vertical_VDC',
       'B_Vertical_CDC', 'B_Vertical_PDC', 'B_Optimal_VDC', 'B_Optimal_CDC',
       'B_Optimal_PDC', 'B_Horizontal_VDC', 'B_Horizontal_CDC',
       'B_Horizontal_PDC', 'B_Vertical_Temperature', 'B_Optimal_Temperature',
       'B_Horizontal_Temperature'])
```
